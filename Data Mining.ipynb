```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Membaca file csv yang kemudian dikelola pada DataFrame
```

```python
df = pd.read_csv('covid19.csv')
df
```

Memfilter setiap kolom bagian Country/Region dengan Indonesia

```python
indonesia = df[df['Country/Region'] == 'Indonesia']
indonesia
```

Memfilter kasus Indonesia yang berjumlah di atas atau lebih dari 0

```python
kasus = indonesia[indonesia['Confirmed']>0]
kasus
```

Visualisasi data dengan matplotlib

```python
plt.plot(kasus['Date'], kasus['Confirmed'])
plt.show()
```

Memfilter kasus untuk bulan Mei

```python
kasusmei = kasus[kasus.Date >= '5/1/20']
kasusmei
```

Membagi setiap kasus di Bulan Mei, kasus yang terkonfirmasi, yang sembuh, dan yang meninggal.

```python
plt.plot(kasusmei.Date, kasusmei.Confirmed)
plt.plot(kasusmei.Date, kasusmei.Recovered)
plt.plot(kasusmei.Date, kasusmei.Deaths)
plt.xticks(rotation = 'vertical')
plt.legend(['Jumlah kasus', 'Sembuh', 'Maninggal'])
plt.show()
```
